```python
from pylibdmtx import pylibdmtx
import time
import cv2
import glob
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
def decode_id(path):
        im=img.imread(path)
        plt.imshow(im,cmap='gray')
        plt.title("title")
        plt.show
        #
        all_barcode_info=pylibdmtx.decode(im,timeout=500,max_count=1)
        if len(all_barcode_info)!=0:
            return (all_barcode_info[0].data.decode("utf-8"))
        else:
            return 0
#         print(all_barcode_info[0].data.decode("utf-8"))
#         if (code.strip()!=""):
#             print(code)

#ttemplate       
file=glob.glob('C:\\Users\\perry\\CV\\ID\\'+'*.jpg')
for i,f in enumerate(file):
    code=decode_id(f)
    if code ==0:
        None
    else:
        print(code)
        break
    
```
